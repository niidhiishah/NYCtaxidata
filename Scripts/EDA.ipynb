```python
# Display basic information about the DataFrame
print("Observations (Records):", len(df))
print("List of Variables (Columns):", df.columns.tolist())
print("Number of Missing Fields:")
print(df.isnull().sum())

# Compute statistics for numeric variables
numeric_stats = df.describe()

# Compute min and max dates for date variables
date_vars = df.select_dtypes(include=['datetime64'])
date_stats = date_vars.agg(['min', 'max'])

# Print statistics for numeric variables and date variables
print("Numeric Variable Statistics:")
print(numeric_stats)
print("Date Variable Statistics:")
print(date_stats)

# Basic graphs/plots with Matplotlib
import matplotlib.pyplot as plt

# Plot a histogram 
# plt.hist(df['tips'], bins = [0,25,50,75,100,125,150,175,200])
plt.hist(df['tips'], bins = [0,25,50,75,100,125,150,175,200])
# Show the plot
plt.show()

# Value counts for categorical variables
categorical_vars = df.select_dtypes(include=['object'])
for col in categorical_vars.columns:
    print(f"Value Counts for {col}:")
    print(df[col].value_counts())
    print()
        
# Distribution of numeric variables
numeric_vars = df.select_dtypes(include=['float64', 'int64'])
for col in numeric_vars.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True) 
    plt.show()
        
# Remove outliers using z-score method
z_scores = stats.zscore(numeric_vars)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df_filtered = df[filtered_entries]
    
# Print statistics for numeric variables after filtering outliers
numeric_stats_filtered = df_filtered.describe()
print("Numeric Variable Statistics (After Filtering Outliers):")
print(numeric_stats_filtered)
```
